```python
from web3 import Web3
import solcx


w3 = Web3(Web3.HTTPProvider("HTTP://127.0.0.1:8545"))

address = "0xC9b11fD56AAa22617AEbF9A9982A826cbF7DE4f1"
pk_key = "a44761d9cdd3724cb110c03048f997ba646f415a277a9d259c42d9a68739acbe"
```

```python
nonce = w3.eth.getTransactionCount(address)
res = solcx.compile_files(
    ["contracts/Auc.sol"],
    output_values=["abi", "bin"],
    solc_version="0.8.0"
)
abi = res['contracts/Auc.sol:AucEngine']['abi']
bytecode = res['contracts/Auc.sol:AucEngine']['bin']
token_contract = w3.eth.contract(abi=abi, bytecode=bytecode)
tx = token_contract.constructor().buildTransaction(
{
    "from": address,
    "nonce": nonce,
    "gasPrice": w3.eth.gas_price,
    "chainId": 1337,
})
signed_txn = w3.eth.account.sign_transaction(tx, private_key=pk_key)
tx_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
address_con = tx_receipt['contractAddress']
contract = w3.eth.contract(abi=abi, address=address_con)
```

```python
nonce = w3.eth.getTransactionCount(address)
tx = contract.functions.createAuction(1000000, 1, "car", 1).buildTransaction(
    {
        "from": address,
        "nonce": nonce,
        "chainId": 1337,
        "gasPrice": 0,
    }
)
signed_txn = w3.eth.account.sign_transaction(tx, private_key=pk_key)
tx_hash = w3.eth.send_raw_transaction(signed_txn.rawTransaction)
tx_receipt = w3.eth.wait_for_transaction_receipt(tx_hash)
```
